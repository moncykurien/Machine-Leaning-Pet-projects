```python
def get_count(dataset,col,symbol,threshold):
    '''
    This function gets the number of rows that are having values in comparision to the threshold
    Symbol has the following:
    eq - equals
    ne - not equals'''
    if symbol == 'eq':
        return len(dataset[dataset[col]==threshold])
    elif symbol == 'ne':
        return len(dataset[dataset[col] != threshold])
```

```python
#There were zeros in the monetory fields which needs to be treated.
#We need to be concerned about the zeros in the budget and the revenue columns which is not goos.
#This can influence the results.
print("Total rows having budget as 0: ",get_count(movies_cleaned,'budget_in_USD','eq',0))
print("Total rows having revenue as 0: ",get_count(movies_cleaned,'revenue_in_USD','eq',0))
print("Total rows having runtime as 0: ",get_count(movies_cleaned,'runtime','eq',0))
```

```python
movies_cleaned.shape
```

Since there is a great percentage of rows having zeros, it is better to drop those instead of trying to replace the zeros with some number. To drop the zeros should be coverted to NANs and then can be dropped.

```python
#Replacing zeros with nans
movies_cleaned['revenue_in_USD'] = movies_cleaned['revenue_in_USD'].replace(0,np.nan)

movies_cleaned.shape
```

```python
def get_high_low(dataset,col):
    '''This gets the details of the highest and the lowest data of the specific column and dataset passed as argument'''
    #getting the value of highest data in col
    index_with_highest_value = dataset[col].idxmax()
    data_from_high_index = pd.DataFrame(movies_cleaned.loc[index_with_highest_value])
    
    #getting the value of lowest data in col
    index_with_lowest_value = dataset[col].idxmin()
    data_from_low_index = pd.DataFrame(movies_cleaned.loc[index_with_lowest_value])
    
    output = pd.concat([data_from_low_index,data_from_high_index], axis = 'columns')
    
    return output
    
```

```python
def draw_hist_graph(dataset,col,x_label,y_label,title,binsize):
    '''This function produces a distribution graph in the form of a histogram'''
    plt.figure(figsize = (15,6),dpi=100)
    plt.hist(dataset[col],rwidth = 0.9, bins=binsize)

    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
```

```python
def draw_line_graph(data,x_label,y_label,title):
    '''
    THis function produces a line graph for the data provided as input
    '''

    plt.plot(data)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.show()
```

Since we can see some very low profits, lets stick to rows with significant profits. Lets look at data with profit more than 500,000,000

```python
#selecting the movies having profit $50M or more
profit_movie = movies_cleaned[movies_cleaned['profit_in_USD'] >= 50000000]

#reindexing new data
profit_movie.index = range(len(profit_movie))

#we will start from 1 instead of 0
profit_movie.index = profit_movie.index + 1

#displaying the top few rows
profit_movie.head(3)
```

```python
def get_count_in_ascend(dataset,col,sepr = '|'):
    '''This function will count the number of times each values in the field occur and 
    returns a Series in ascending oder per the count'''
    data = dataset[col].str.cat(sep = sepr)
    data = pd.Series(data.split(sepr))
    return data.value_counts(ascending = False)
```

```python
#This helps us find the genres that are popular
genre_count = get_count_in_ascend(profit_movie,'genres')
print(genre_count)
```

```python
#Plotting a horizontal bar chart with genres to see which genres are mostly used

genre_sort = genre_count.sort_values(ascending = True, inplace = True)
#initializing plot
ax = genre_count.plot.barh()

#giving a title
ax.set(title = 'Genres that are filmed the most')

#x-label
ax.set_xlabel('Number of Movies filmed in the genre')

#giving the figure size(width, height)
ax.figure.set_size_inches(15, 8)

#Displaying the plot
plt.show()
```

Most of the movies that made profit more than 50M USD are of the genre 'comedy'. Following the comedy genre, Drama,Action and Thriller genre also are toping the list.The genres on the bottom are War,History,Western,Documentary and Foreign.<br> 
However this only proves that most of the movies in the $50M profit list are in Comedy,Drama,Action and Thriller, but does not prove that these genres guarantee big profits. These genres seem to attract significant interest of the audience.

I would like to know movies released in which month has made the highest profits.

```python
#Making a copy of the release_date and profit columns into a new dataframe
req_list = ['release_date','profit_in_USD']
profit_release = profit_movie[req_list].copy()
profit_release.head(3)
```

```python
#To get the months with more profits, we need to group all of the months of years and then calculate the profits of those
#months giving a new dataframe which gives 'release-date' as index
profit_release = profit_movie.set_index('release_date')

#group all the data by month, since release date is in form of index, we extract month from it
groupby_index = profit_release.groupby([(profit_release.index.month)])

#Calculating the total profit per month
profit_per_month = groupby_index['profit_in_USD'].sum()

#converting table to a dataframe
profit_per_month= pd.DataFrame(profit_per_month)

#creating a list of months
month_list = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']


monthly_movie_profit = sns.barplot(x = profit_per_month.index, y = profit_per_month['profit_in_USD'], data = profit_per_month)
#setting size of the graph
monthly_movie_profit.figure.set_size_inches(15,8)

#setting labels
monthly_movie_profit.set_xlabel("Months", color="g", fontsize = 15)
monthly_movie_profit.set_ylabel("Total Profit", color="r", fontsize = 25)

#setting the title and customizing
monthly_movie_profit.axes.set_title('Total profit per month', color="b", fontsize = 15, alpha = 0.6)

#customizing axes values
monthly_movie_profit.tick_params(labelsize = 5, labelcolor="black")

#rotating the x-axis values to make it readable
monthly_movie_profit.set_xticklabels(month_list, rotation = 40, size = 15)

#shows the plot
plt.show()
```

From the above plot we can find out that the month June has yielded the most profit in the movie industry followed by December and then May.
